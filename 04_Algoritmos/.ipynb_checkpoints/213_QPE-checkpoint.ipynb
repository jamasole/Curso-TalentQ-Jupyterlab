<table width="100%">

<b> Estimación de Fase Cuántica </b>
</table>   

 Supongamos que un  operador unitario $U$ tiene un autovector $\ket{u}$. Por ser unitario, su autovalor es una fase compleja 
<br>

$$
U \ket{u} = e^{2\pi i \varphi} \ket{u}\, ,
$$
<br>

**Promesa**: podemos preparar  $\ket{u}$ y aplicar el operador $U$ tantas veces como queramos. 


**Problema**:  calcular  la *mejor aproximación posible* a $\varphi$. 

 
Las fases  $e^{i2\pi \varphi}=e^{i2\pi (\varphi+n)}$  son iguales, para cualquier $n\in {\mathbb Z}$ entero. Por tanto será suficiente considerar  $\varphi \in [0,1)$ para tener todos los *autovalores distintos* posibles.
 
 
 

<center> 
<img src="./figuras/QPE.png" width='50%' style='margin:automatic'/>
</center>


El circuito  QPE ("Quantum Phase Estimation") viene representado en la figura,  donde el primer registro tiene dimensión $t$ y la entrada es el estado $\ket{0}^{\otimes t} \ket{u}$.  

La anchura $t$ **controlará la precisión de nuestra aproximación** a $\varphi$.  La dimensión del espacio al que pertenece $\ket{u}$  es la necesaria para servir de espacio de representación para $U$.



Es fácil convencerse con un par de ejemplos que el efecto de la parte central (controlada) del circuito representado 
es llevar el estado $\ket{y}\ket{u} \to \ket{y} U^y\ket{u}$, y que esto no depende de que $\ket{u}$ sea un autoestado. 


Cuando $\ket{u}$ es un autoestado, se produce un retroceso de fase y la salida de la parte central de este circuito  es
<br>
<br>

\begin{eqnarray}
\ket{\psi} &=& \frac{1}{2^{t/2}} \sum_{y} \ket{y} U^{y}\ket{u} \\ \rule{0mm}{10mm}
&=&  \frac{1}{2^{t/2}} \sum_{y} e^{2\pi i \varphi y} \ket{y} \otimes \ket{u} \nonumber\\  \rule{0mm}{10mm}
&=& \frac{1}{2^{t/2}} \sum_{y} e^{2\pi i (2^t \varphi)  y/2^t} \ket{y} \otimes \ket{u}  \label{fase1}  
\end{eqnarray}


El estado en el primer registro ¡**parece** la  transformada de Fourier!  
de un estado $\ket{2^t \varphi}\in \Hil^{\otimes 2^t}$. 



$$
\ket{\widetilde{2^t \varphi}}  =  \frac{1}{2^{t/2}} \sum_{y} e^{2\pi i (2^t \varphi)  y/2^t} \ket{y}
$$



Sin embargo $2^t\varphi$ en general *no será
un entero*.

 Vamos entonces a separarlo en la forma siguiente en sus *partes entera* y *mantisa*

$$
2^t \varphi  =a + \delta \label{repfase}
$$

con $\delta \in [0,1)$.
 
 


<div class="alert alert-block alert-warning">
<b> Ejemplo: </b>

Si $\varphi = 0.4375$ Entonces 
- $t = 2 \Rightarrow 2^t \varphi = 1.75 \Rightarrow \delta = 0.75$ 
- $t = 3 \Rightarrow 2^t \varphi = 3.5~ \Rightarrow \delta = 0.5$ 
- $t = 4 \Rightarrow 2^t \varphi = 7  ~~~ \Rightarrow \delta = 0$
    
En el fondo esto es así porque $0.4375 = 2^{-2} + 3 * 2^{-4}$    
</div>

<div class="alert alert-block alert-info">
<p style="text-align: center;"> <p style="text-align: left ;color: navy;">  
<b>Definición:</b> 
<br> 
decimos que $a = [2^t\varphi]$ es un  <b>estimador de $\varphi$ de orden $t$</b> si la distancia  
<br> 
<br>
$$|\varphi - a/2^t| = \delta/2^t 
$$ 
<br>    
disminuye (exponencialmente)  con la anchura $t$ del primer registro. 
</div>

Introduciendo este cambio de variable  podemos ahora efectuar la transformada de Fourier inversa en el primer registro
<br>

$$
U^\dagger_{TFC} \frac{1}{\sqrt{2^t}} \sum_{y=0}^{2^t-1} e^{2\pi i  \, (a + \delta) y/2^t} \ket{y} ~=~ \frac{1}{2^{t}} \sum_{x=0}^{2^t-1} \sum_{y=0}^{2^t-1}  e^{2\pi i (\delta + a -x) y/2^t} \ket{x}  = \ket{\psi}
$$

<br>

-  Si  $\delta = 0$, es decir, si $2^t\varphi$ es entero, entonces el resultado
es exactamente $\ket{\psi} = \ket{a}$ 
 
 $$
 U^\dagger_{FT}(U^{2^{t-1}}U^{2^{t-2}}... U^{2^0} H^{\otimes t}\ket{0} )=  U^\dagger_{FT}\ket{\widetilde{2^t\varphi}} = U^\dagger_{FT}  U_{FT}\ket{a} = \ket{a} \, .
 $$
     


 Midiendo el primer registro,  recuperamos el registro binario del número $a\in[0, 2^t-1]$ y de ahí, recuperamos la fase buscada 
 
 $$
 \varphi = \frac{a}{2^t}  \in [0,1-2^{-t}]$$ 
 
 de forma *exacta*.


- Si $\delta \neq 0$ entonces una medida del primer registro  dará el registro binario de un número $m \in [0,2^t-1)$ que, ahora no será $m\neq a$. 

De no ser $2^t\theta$ entero, la TFC inversa devolverá *una combinación lineal* de estados con  amplitudes de probabilidad centradas en $x=2^t\theta$, por lo que esos estados serán medidos con una mayor probabilidad (pero no del 100%).


Vamos a estudiar esta distribución analizando el operador unitario de fase

$$
U = P(\phi) = \begin{bmatrix} 1 & 0 \\ 0 & e^{i\phi}\end{bmatrix} ~~~~\Rightarrow ~~~~P(\phi) \ket{1} = e^{i\phi}\ket{1}
$$

por tanto $\ket{1} = \ket{u}$ es el autoestado. 

Para poner a prueba el algoritmos vamos a  por seleccionar un valor de $\phi$ de la forma  

$$\phi = (a + \delta) 2^{-t}$$

```python
# Select a phase phi and a number of qubits for the representation space
t_here=3
a = 3.
delta = 0.1
phi=(a+ delta)*2.**(-3) 

print('phi=', phi, ' ->   2^t*phi = a + delta = ', 2**t_here*phi , '  ,   a =' ,int(2**t_here*phi), '=', format(int(2**t_here*phi),'b').zfill(t_here) )
```

```python

t_circuit = 6
n_qubits= t_circuit 
n_bits=n_qubits

qr=QuantumRegister(n_qubits,name="q")
ar=QuantumRegister(1,name="ancilla")
cr=ClassicalRegister(n_bits)

qpe = QuantumCircuit(qr,ar,cr)   

qpe.x(ar[0]) #  inicializamos la ancilla al autoestado de P(phi) ->  |u> = |1>
#qpe.h(ar[0]) #  

qpe.h(range(n_qubits))
qpe.barrier()

# aplicamos los operadores controlados c-U^j 
for j in range(n_qubits):
    qpe.cp(2*np.pi*phi*(2**j), j, n_qubits); # This is C-U

qpe.barrier()

# we apply the inverse Fourier transform from qiskit.circuit.library. 
# for some reason I don't understant it does not work with the swap inside the qft.dagger 

qft_dagger(qpe, n_qubits)

# but a swap needs to be inserted to get the correct readout
#swap_registers(qpe, n_qubits)

qpe.measure(qr,cr)

qpe.draw(output="mpl")
```

```python
from qiskit import Aer, execute
M_backend = Aer.get_backend('qasm_simulator')

from qiskit.visualization import plot_histogram
results = execute(qpe, M_backend, shots=2048).result()
counts = results.get_counts()
plot_histogram(counts)
```

De aquí extraemos la aproximación que hemos obtenido. 

```python
max(counts, key=counts.get)
```

```python
a_resultante = int(max(counts, key=counts.get),2)

print('el valor de phi original era =', phi)

print('la mejor aproximación al valor verdadero es a/2^t =', a_resultante*2**(-t_circuit))

print('el error relativo es delta/2^t/phi =', (phi -a_resultante*2**(-t_circuit))/phi)
```

De hecho, distintas medidas, del estado resultante $\ket{m}= \ket{a+l}$ se distribuirán en el torno a  $\ket{a} = \ket{[2^t\varphi]} $.

 Para una mantisa fija, $\delta = \phi - a$, podemos obtener la *probabilidad de obtener* un estado $\ket{a+l}$
<br>
<br>

\begin{eqnarray}
p(\delta, l) ~~&=&~~ \left\vert \,  \bra{a+ l} \,  \frac{1}{2^{t}}\sum_{x=0}^{2^t-1}   \sum_{y=0}^{2^t-1}  e^{2\pi i (\delta + a -x) y/2^t} \ket{x}  \,  \right\vert^2  ~~ \stackrel{x\to a + l}{=} ~~
  \frac{1}{2^{2t}} \left\vert \,  \sum_{y=0}^{2^t-1} \left(e^{2\pi i (\delta-l) /2^t}\right)^y \,  \right\vert^2 
\end{eqnarray}

\begin{eqnarray}
&=& ~~  \frac{1}{2^{2t}} \left\vert\,  \frac{1- e^{2\pi i  \,  (\delta-l) }}{1- e^{2\pi i (\delta-l)/2^t}} \, \right\vert^2   ~= ~ 
\frac{1}{2^{2t}} \left\vert \, \frac{\sin \pi   (\delta-l)}{\sin (\pi(\delta-l)/2^t)} \, \right\vert^2 \, 
\hspace{1.3cm}
\end{eqnarray}


$$
\geq   \frac{1}{2^{2t}} \left(\frac{2 (\delta-l)}{\pi(\delta-l)/2^t }\right)^2 = \left(\frac{2}{\pi}\right)^2 \sim 0.405 \hspace{3.8cm}
$$
<br>
<br>

donde hemos usado que   $\pi x\geq \sin(\pi x) \geq 2 x$ para $x\in [ 0,\pi/2 ]$.
