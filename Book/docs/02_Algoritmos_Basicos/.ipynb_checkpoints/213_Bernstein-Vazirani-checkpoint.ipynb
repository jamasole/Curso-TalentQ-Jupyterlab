## Introducción

Sea $f(x):\{0,1\}^n \rightarrow \{0,1\}$ una función tal que $f(x)=c·x = c_0x_0\oplus c_1x_1 \oplus ... \oplus c_{n-1}x_{n-1}$, donde $c$ es una cadena de $n$ bits. Queremos encontrar $c$ con una sola medida. Nótese que hay $2^n$ valores posibles.

## Algoritmo
El algoritmo de Bernstein-Vazirani sigue el mismo procedimiento que el de Deutsch-Josza. La diferencia está en el último paso. El estado que tenemos antes de la medida final en Deutsch-Josza es el siguiente:

\begin{equation}
{\frac{1}{2^n}\sum_{k=0}^{2^n-1}\left[\sum_{x=0}^{2^n-1}(-1)^{f(x)+k·x}\right]\left|k\right>}.
\end{equation}


Si incluimos la condición $f(x)=c·x$ tenemos:

\begin{equation}
{\frac{1}{2^n}\sum_{k=0}^{2^n-1}\left[\sum_{x=0}^{2^n-1}(-1)^{c·x+k·x}\right]\left|k\right>},
\end{equation}

que podemos demostrar que es igual al estado $\left|c_{n-1}...c_1c_0\right>$, de modo que una medida sobre los $n$ qubits del registro nos da directamente el valor de $c$. Para ello, solo hay que trabajar un poco la expresión. Si tenemos en cuenta que

i) $(-1)^a = (-1)^{a (mod 2)}$, y

ii) $(a+b) (mod 2) = (a (mod 2) + b (mod 2)) (mod 2)$;

podemos demostrar que:

\begin{equation}
{
    (-1)^{c·x+k·x}=(-1)^{(c_0+k_0)x_0 \oplus ... \oplus (c_{n-1}+k_{n-1})x_{n-1}} = (-1)^{(c \oplus k)·x}.
}
\end{equation}

Aquí, el exponente cumple la siguiente propiedad: $c=k \Rightarrow c \oplus k = 0$. Como consecuencia, tenemos el siguiente resultado:

\begin{equation}
{
    \frac{1}{2^n}\sum_{x=0}^{2^n-1}(-1)^{(c\oplus k)·x} = \left\{\begin{array}{cl} 0 &\hbox{si $c \neq k$}\\
1 & \hbox{si $c = k$} 
\end{array}
\right.
}
\end{equation}

ya que, cuando $c\neq k$, la mitad de sumandos valdrán $+1$ y la otra mitad $-1$ debido a que estamos haciendo todas las combinaciones posibles con $x$. En cambio, todos valdrán $+1$ cuando $c=k$ por la condición anterior.

## Implementación

En Qiskit, podemos probar este algoritmo del siguiente modo.

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit

def BV_circuit(n, c):
    #n: número de bits
    #c: pendiente de la función
    #return: circuito

    qreg = QuantumRegister(n+1)
    creg = ClassicalRegister(n)
    qc = QuantumCircuit(qreg,creg)

    #Hacemos máxima superposición
    qc.h(qreg)

    #Ponemos el último qubit en el estado |->
    qc.z(qreg[-1])

    qc.barrier()
    #Construimos la función lineal del siguiente modo

    if type(c) is not str:

        c = str(c)

    #Como las posiciones de los bits se cuentan en sentido contrario
    #a las de los qubits, usamos un contador

    counter = n
    for bit in c:

        counter -= 1

        if bit == '1':

            qc.cx(qreg[counter],qreg[-1])

    qc.barrier()

    #Aplicamos Hadamard de nuevo

    qc.h(qreg[0:-1])

    qc.measure(qreg[0:-1],creg)

    return qc
```

```python
circuito = BV_circuit(6,110010)

circuito.draw(output = 'mpl')
```

```python
#Comprobamos que se obtiene c al medir

from qiskit import execute, Aer

simulator = Aer.get_backend('qasm_simulator')

job = execute(circuito,simulator,shots = 1000)

counts = job.result().get_counts(circuito)

print(counts)
```
